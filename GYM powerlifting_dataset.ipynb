```python
plt.figsize = (50,50)
sns.histplot(data ,y="Amount Lifted (kg)",x="Age")
```

```python
plt.figsize = (1000,1000)
sns.displot(data, y="Age",x="Lift Type")
```

```python
plt.figsize = (10,20)
sns.relplot(data ,y="Weight Class", x="Amount Lifted (kg)")
```

```python
plt.figsize = (100,100)
sns.barplot(data , y="Lifter Name", x="Age")
```

```python
sns.lmplot(y="Amount Lifted (kg)", x="Age", data=data)
```

```python
plt.figsize = (100,100)
sns.catplot(data=data, y="Weight Class", x="Lift Type")
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 10))
sns.histplot(data=data, x='Lift Type', hue='Amount Lifted (kg)', bins=10, kde=True, multiple="stack")

# Add labels
plt.xlabel('Lift Type')
plt.ylabel('Amount Lifted (kg)')

# Show the plot
plt.show()
```

```python
status_counts = data['Weight Class'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
sns.boxplot(x="Age", y="Weight Class", data=data)

plt.xlabel("Age")
plt.ylabel("Weight Class")
plt.show()
```

```python
sns.violinplot(x="Weight Class", y="Lift Type", data=data)

plt.xlabel("Weight Class")
plt.ylabel("Lift Type")
plt.show()
```

```python
corr_matrix = data.select_dtypes(include=['number']).corr()

# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True), linewidths=0.5)

plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['number']))

plt.show()
```
